```python
df.columns=df.columns.str.replace(" ","")
df
```

```python
df = df.rename(columns={'TotalCost(Current)': 'Totalcost'})
df
```

```python
df.UnitCost=df.UnitCost.replace('[$]','',regex=True)
df
```

```python
df.Totalcost=df.Totalcost.replace('[$,]','',regex=True)
df
```

```python
df.Current=df.Current.fillna(0)
df.Future=df.Future.fillna(0)
df.Historic=df.Historic.fillna(0)
df.UnitCost=df.UnitCost.fillna(0)
df.Totalcost=df.Totalcost.fillna(0)
df.Orders=df.Orders.fillna(0)
```

```python
df[df.AircraftType=='Airbus A319']
```

```python
df['Total']=df['Total'].fillna(df[['Current','Future','Historic']].sum(axis=1))
df
```

```python
df['AverageAge'] = df['AverageAge'].fillna(df['AverageAge'].mean())
df
```
